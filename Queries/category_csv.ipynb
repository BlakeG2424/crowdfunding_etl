```python
import pandas as pd
```

```python
# Read the data into a Pandas DataFrame
crowdfunding_info_df = pd.read_excel('Resources/crowdfunding.xlsx')
crowdfunding_info_df.head()
```

```python
crowdfunding_info_df[['category', 'sub-category']] = crowdfunding_info_df['category & sub-category'].str.split('/', expand=True)
crowdfunding_info_df
```

```python
# Get the unique categories from the original DataFrame
unique_categories = crowdfunding_info_df['category'].unique()

# Create the Category DataFrame
category_df = pd.DataFrame({
    'category_id': range(1, len(unique_categories) + 1),
    'category': unique_categories
})

# Print the Category DataFrame
category_df
```

```python
#Export the DataFrame as a 'category.csv' CSV file

category_df.to_csv('category.csv', index=False)
```

```python
# Get the unique categories from the original DataFrame
unique_subcategory = crowdfunding_info_df['sub-category'].unique()

# Create the Category DataFrame
sub_category_df = pd.DataFrame({
    'category_id': range(1, len(unique_subcategory) + 1),
    'category': unique_subcategory
})

# Print the Category DataFrame
sub_category_df
```

```python
#Export the DataFrame as a 'subcategory.csv' CSV file

sub_category_df.to_csv('subcategory.csv', index=False)
```
